```python
%matplotlib inline

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import pickle
from sklearn.linear_model import Ridge
from tqdm.notebook import tqdm
import random
```

```python
df = pd.read_csv('data/allLanguages_regression_model_averageLanguageStability_crossValidation_errors.csv')
df = df.drop("Unnamed: 0",axis=1)
```

```python
#Avg. error of regression model
print(sum(df.error)/len(df))
print(np.std(df.error))
```

```python
#What if we just guessed the average ground truth instead?
errors = []
avg_guesses = []
for language in list(df.language):
    df_mod = df.loc[df.language != language]
    avg_guess = sum(df_mod.ground_truth)/len(df_mod)
    df_ground_truth = list(df.loc[df.language==language]['ground_truth'])[0]
    error = abs(avg_guess-df_ground_truth)
    errors.append(error)
    avg_guesses.append(avg_guess)
    
```

```python
print(sum(errors)/len(errors))
```

```python
np.std(errors)
```

```python
df['prediction_error'] = df.error
df = df.drop('error',axis=1)
df['avg_guess'] = avg_guesses
df['avg_guess_error'] = errors
```

```python
df = df[['language','ground_truth','avg_guess','avg_guess_error']]
```

```python
oneFeature = pd.read_csv('data/crossValidation_oneFeature.csv')
```

```python
oneFeature['oneFeature_prediction'] = oneFeature.prediction
oneFeature['oneFeature_error'] = oneFeature.error
oneFeature = oneFeature[['language','oneFeature_prediction',\
                         'oneFeature_error']]
```

```python
lowReg = pd.read_csv('data/crossValidation_lowRegularization.csv')
highReg = pd.read_csv('data/crossValidation_highRegularization.csv')
lowReg['lowReg_prediction'] = lowReg.prediction
highReg['highReg_prediction'] = highReg.prediction
lowReg['lowReg_error'] = lowReg.error
highReg['highReg_error'] = highReg.error
lowReg = lowReg[['language','lowReg_prediction','lowReg_error']]
highReg = highReg[['language','highReg_prediction','highReg_error']]
```

```python
df = df.merge(oneFeature,on='language')
df = df.merge(lowReg,on='language')
df = df.merge(highReg,on='language')
```

```python
df = df.sort_values(by='ground_truth',ascending=False)
```

```python
plt.figure(figsize=(10,5))
plt.bar([i for i in range(len(df))],df.ground_truth,label='Ground truth')
plt.plot([i for i in range(len(df))],df.prediction,label='Predicted',\
         color='r')
plt.plot([i for i in range(len(df))],df.avg_guess,label='Avg. guess',\
        color='g',linestyle='dashed')
#plt.plot([i for i in range(len(df))],df.oneFeature_prediction,\
#         label='Predicted (one feature)',color='k')
#plt.plot([i for i in range(len(df))],df.lowReg_prediction,\
#        label='Predicted (regularization=0.0001)',color='m')
plt.plot([i for i in range(len(df))],df.highReg_prediction,\
        label='Predicted (regularization=100)',color='k')
plt.ylabel('Avg. Stability (%)')
plt.xlabel('Languages')
plt.legend()
```
