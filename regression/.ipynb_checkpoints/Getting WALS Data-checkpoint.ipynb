```python
%matplotlib inline

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from tqdm import tqdm_notebook
from collections import Counter
```

```python
wals_path = '../../embedding_datasets/wals/wals_dataset/'
```

```python
languages = pd.read_csv(wals_path+'languages.csv')
```

```python
parameters = pd.read_csv(wals_path+'parameters.csv')
```

```python
values = pd.read_csv(wals_path+'values.csv')
```

```python
parameters
```

```python
polyglot_languages = ['ar','bg','ca','cs','da','de','el','en','es','et','fa','fi','fr','he','hi','hr','hu','id',
                      'it','ja','ko','lt','lv','ms','nl','no','pl','pt','ro','ru','sk','sl','sr','sv','th','tl','tr','uk',
                      'vi','zh']
polyglot_languages_long = ['Arabic','Bulgarian','Catalan','Czech','Danish','German','Greek, Modern','English',
                           'Spanish; Castilian','Estonian','Persian','Finnish','French','Hebrew (modern)','Hindi',
                           'Croatian','Hungarian','Indonesian','Italian','Japanese','Korean','Lithuanian',
                           'Latvian','Malay','Dutch','Norwegian','Polish','Portuguese',
                           'Romanian, Moldavian, Moldovan','Russian','Slovak','Slovene','Serbian','Swedish','Thai',
                           'Tagalog','Turkish','Ukrainian','Vietnamese','Chinese']
```

```python
wals_languages_long = ['Arabic (Modern Standard)','Bulgarian','Catalan','Czech','Danish','German','Greek (Modern)','English',
                           'Spanish','Estonian','Persian','Finnish','French','Hebrew (Modern)','Hindi',
                           'Serbian-Croatian','Hungarian','Indonesian','Italian','Japanese','Korean','Lithuanian',
                           'Latvian','Malay','Dutch','Norwegian','Polish','Portuguese',
                           'Romanian','Russian','Slovak','Slovene','Serbian-Croatian','Swedish','Thai',
                           'Tagalog','Turkish','Ukrainian','Vietnamese','Cantonese']
wals_languages = [list(languages.loc[languages.Name==language_name]['ID'])[0] for language_name in wals_languages_long]
```

```python
parameter_ids = list(parameters['ID'])
```

```python
all_data = {}
for parameter_id in tqdm_notebook(parameter_ids):
    parameters = []
    for language in wals_languages:
        lookup = values.loc[(values.Language_ID==language) & (values.Parameter_ID==parameter_id)]
        if len(lookup)==0:
            parameters.append('')
        else:
            parameters.append(list(lookup['Value'])[0])
    all_data[parameter_id] = parameters
```

```python
#all_data['polyglot_language_short'] = polyglot_languages
#all_data['polyglot_language_long'] = polyglot_languages_long
#all_data['wals_language_short'] = wals_languages
#all_data['wals_language_long'] = wals_languages_long
```

```python
results = pd.DataFrame(data=all_data)
```

```python
current_columns = results.columns.values
```

```python
parameters = pd.read_csv(wals_path+'parameters.csv')
```

```python
new_columns = [i+': '+str(list(parameters.loc[parameters.ID==i]['Name'])[0]) for i in current_columns]
```

```python
results.columns = new_columns
```

```python
results
```

```python
results['polyglot_language_short'] = polyglot_languages
results['polyglot_language_long'] = polyglot_languages_long
results['wals_language_short'] = wals_languages
results['wals_language_long'] = wals_languages_long
```

```python
results.to_csv('wals_values.csv')
```

```python
f, axarr = plt.subplots(11,11,figsize=(100,100),sharey=True)
row = 0
column = 0
for parameterName in results.columns.values[:-4]:
    counts = Counter([i if i !='' else 'Unknown' for i in results[parameterName]])
    if counts['Unknown'] > len(results[parameterName])/2: #If we don't know 50% or more of the values, ignore
        continue
    df = pd.DataFrame.from_dict(counts, orient='index')
    df.plot(kind='bar',ax=axarr[row,column])
    axarr[row,column].set_title(parameterName)
    axarr[row,column].get_legend().remove()
    
    column += 1
    if column > 10:
        column = 0
        row += 1
plt.savefig('visualizations/all_wals.pdf',dpi=100)
```
