```python
import pandas as pd
```

```python
wals = pd.read_csv('wals_values_bible.csv')
wals = wals.drop("Unnamed: 0",axis=1)
```

```python
#https://thispointer.com/how-to-find-drop-duplicate-columns-in-a-dataframe-python-pandas/
def getDuplicateColumns(df):
    '''
    Get a list of duplicate columns.
    It will iterate over all the columns in dataframe and find the columns whose contents are duplicate.
    :param df: Dataframe object
    :return: List of columns whose contents are duplicates.
    '''
    duplicateColumnNames = set()
    # Iterate over all the columns in dataframe
    for x in range(df.shape[1]):
        # Select column at xth index.
        col = df.iloc[:, x]
        # Iterate over all the columns in DataFrame from (x+1)th index till end
        for y in range(x + 1, df.shape[1]):
            # Select column at yth index.
            otherCol = df.iloc[:, y]
            # Check if two columns at x 7 y index are equal
            if col.equals(otherCol):
                duplicateColumnNames.add(df.columns.values[y])
 
    return list(duplicateColumnNames)
```

```python
duplicated = getDuplicateColumns(wals) #all nan in the column right now
```

```python
wals = wals.drop(duplicated,axis=1)
```

```python
getDuplicateColumns(wals)
```

```python
wals_value_names = wals.columns.values[:-4]
```

```python
wals_value_names
```

```python
wals = wals.fillna('')
```

```python
new_value_names = []
for name in wals_value_names:
    print(name)
    values = [i for i in list(set(list(wals[name])))]
    print(values)
    for value in values:
        wals[name+'__'+str(value)] = [1 if i==value else 0 for i in wals[name] ]
        new_value_names.append(name+'__'+str(value))
```

```python
wals_binary = wals[new_value_names+['language']] #polyglot_language_short
```

```python
wals_binary.to_csv('wals_bible_binary_values.csv')
```
