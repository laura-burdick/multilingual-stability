```python
from sklearn.linear_model import Ridge
import pickle
import pandas as pd
```

```python
dataPath = '../embedding_datasets/language_regression/'
```

```python
with open(dataPath+'regression_model.pkl','rb') as pickleFile:
    model = pickle.load(pickleFile)
```

```python
with open(dataPath+'regression_model_averageLanguageStability.pkl','rb') as pickleFile:
    model_language = pickle.load(pickleFile)
```

```python
with open(dataPath+'regression_model_goodWals.pkl','rb') as pickleFile:
    model_good = pickle.load(pickleFile)
with open(dataPath+'regression_model_averageLanguageStability_goodWals.pkl','rb') as pickleFile:
    model_language_good = pickle.load(pickleFile)
```

```python
wals = pd.read_csv('wals_binary_values.csv')
wals = wals.drop('Unnamed: 0',axis=1)
```

```python
labels = list(wals.columns.values)[:-1]
```

```python
labels.append('Size of Wikipedia')
labels += ['pos_', 'pos_adverb',
       'pos_letter', 'pos_syllable', 'pos_symbol', 'pos_determiner',
       'pos_noun', 'pos_article', 'pos_phrase', 'pos_proverb', 'pos_verb',
       'pos_preposition', 'pos_numeral', 'pos_initialism',
       'pos_adjective', 'pos_proper noun', 'pos_interjection',
       'pos_pronoun', 'pos_conjunction']
labels.append('Ranked frequency')
labels.append('Normalized frequency')
labels.append('Normalized word length')
```

```python
labels_language = list(wals.columns.values)[:-1] + ['Size of Wikipedia']
```

```python
with open('good_wals_values.pkl','rb') as pickleFile:
    labels_good = pickle.load(pickleFile)
labels_good.append('Size of Wikipedia')
labels_good += ['pos_', 'pos_adverb',
       'pos_letter', 'pos_syllable', 'pos_symbol', 'pos_determiner',
       'pos_noun', 'pos_article', 'pos_phrase', 'pos_proverb', 'pos_verb',
       'pos_preposition', 'pos_numeral', 'pos_initialism',
       'pos_adjective', 'pos_proper noun', 'pos_interjection',
       'pos_pronoun', 'pos_conjunction']
labels_good.append('Ranked frequency')
labels_good.append('Normalized frequency')
labels_good.append('Normalized word length')
```

```python
with open('good_wals_values.pkl','rb') as pickleFile:
    labels_language_good = pickle.load(pickleFile)
labels_language_good.append('Size of Wikipedia')
```

```python
weights = model.coef_
weights_language = model_language.coef_
weights_good = model_good.coef_
weights_language_good = model_language_good.coef_
```

```python
model_weights = pd.DataFrame(data={'label':labels,'Regular Model':weights})
```

```python
model_weights = model_weights.sort_values(by='Regular Model')
```

```python
weight_language = []
for it,row in model_weights.iterrows():
    lable = row['label']
    if lable in labels_language:
        weight_language.append(weights_language[labels_language.index(lable)])
    else:
        weight_language.append('nan')
```

```python
model_weights['Average Language Stability Model'] = weight_language
```

```python
weight_good = []
for it,row in model_weights.iterrows():
    lable = row['label']
    if lable in labels_good:
        weight_good.append(weights_good[labels_good.index(lable)])
    else:
        weight_good.append('nan')
model_weights['Regular Model (Good WALS)'] = weight_good
```

```python
weight_good = []
for it,row in model_weights.iterrows():
    lable = row['label']
    if lable in labels_language_good:
        weight_good.append(weights_good[labels_language_good.index(lable)])
    else:
        weight_good.append('nan')
model_weights['Average Language Stability Model (Good WALS)'] = weight_good
```

```python
model_weights.to_csv('model_weights.csv')
```
